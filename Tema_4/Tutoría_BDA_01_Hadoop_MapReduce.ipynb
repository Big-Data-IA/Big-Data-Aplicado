# Ejemplo de Map/Reduce

*WordCountSimple* es una sencilla aplicación que cuenta el número de ocurrencias de cada palabra en un conjunto de entrada dado.

Funciona con una instalación Hadoop local-standalone, pseudo-distribuida o totalmente distribuida.

Más detalles de la solución en este [enlace](https://hadoop.apache.org/docs/stable/hadoop-mapreduce-client/hadoop-mapreduce-client-core/MapReduceTutorial.html#Source_Code)

## Generar el fichero dentro del propio colab

Importación de librerías necesarias

- org.apache.hadoop.mapreduce: clases Mapper y Reducer
- org.apache.hadoop.mapreduce.lib.input: flujos de entrada
- org.apache.hadoop.mapreduce.lib.output: flujos de salida


```python
%%writefile -a WordCountSimple.java
import java.io.IOException;
import java.util.*;

import org.apache.hadoop.mapreduce.*;
import org.apache.hadoop.mapreduce.lib.input.*;
import org.apache.hadoop.mapreduce.lib.output.*;
import org.apache.hadoop.conf.*;
import org.apache.hadoop.io.*;
import org.apache.hadoop.util.*;
import org.apache.hadoop.fs.*;


public class WordCountSimple {
```

La implementación del mapeador, a través del método *map*, procesa una línea a la vez, tal y como lo proporciona el TextInputFormat especificado. A continuación, divide la línea en tokens separados por espacios en blanco, a través del StringTokenizer, y emite un par clave-valor de $< <palabra>, 1>$.

```python
%%writefile -a WordCountSimple.java

public static class Map extends Mapper<LongWritable, Text, Text, IntWritable> {	
		private final static IntWritable one = new IntWritable(1);
		private Text word = new Text();
		
		@Override
		public void map(LongWritable key, Text value, Context context)
				throws IOException, InterruptedException {
			String line = value.toString();
			StringTokenizer tokenizer = new StringTokenizer(line);
			while (tokenizer.hasMoreTokens()) {
				word.set(tokenizer.nextToken());
				context.write(word, one);
			}
		}
	}
```

WordCount también especifica un combinador. Por lo tanto, la salida de cada mapa se pasa a través del combinador local (que es el mismo que el Reductor según la configuración del trabajo) para la agregación local, después de ser ordenado en las claves.

La implementación del Reducer, a través del método reduce, sólo suma los valores, que son los recuentos de ocurrencia para cada clave (es decir, palabras en este ejemplo).



```python
%%writefile -a WordCountSimple.java

public static class Reduce extends Reducer<Text, IntWritable, Text, IntWritable> {
		@Override
		public void reduce(Text key, Iterable<IntWritable> values, Context context)
				throws IOException, InterruptedException {
			int sum = 0;
			for (IntWritable val: values) {
				sum += val.get();
			}
			context.write(key, new IntWritable(sum));
		}

	}
```

El método *main* especifica varias facetas del trabajo, como las rutas de entrada/salida (pasadas a través de la línea de comandos), los tipos de clave/valor, los formatos de entrada/salida, etc., del trabajo. Luego llama a *job.waitForCompletion* para enviar el trabajo y monitorear su progreso.

```python
%%writefile -a WordCountSimple.java

public static void main(String[] args) throws Exception {
		Job job = Job.getInstance();
		job.setJarByClass(WordCountSimple.class);
		job.setJobName("WordCountSimple");
		
		job.setOutputKeyClass(Text.class);
		job.setOutputValueClass(IntWritable.class);

		job.setMapperClass(Map.class);
		job.setCombinerClass(Reduce.class);
		job.setReducerClass(Reduce.class);
		
		job.setInputFormatClass(TextInputFormat.class);
		job.setOutputFormatClass(TextOutputFormat.class);
		
		FileInputFormat.addInputPath(job, new Path(args[0]));
		FileOutputFormat.setOutputPath(job, new Path(args[1]));
		
		System.exit(job.waitForCompletion(true)? 0 : 1);
	}
```

## Hadoop Streaming

Hadoop streaming permite crear y ejecutar trabajos Map/Reduce con cualquier ejecutable o script como mapeador y/o reductor.

Mas información sobre Map/Reduce en el siguiente [enlace](https://hadoop.apache.org/docs/current/hadoop-mapreduce-client/hadoop-mapreduce-client-core/MapReduceTutorial.html)

```python
!find / -name 'hadoop-streaming*.jar'
```

Planteamos la ejecución en streaming

```python
!hadoop jar /usr/local/hadoop-3.3.1/share/hadoop/tools/lib/hadoop-streaming-3.3.1.jar
```

```python
%%bash
hadoop jar /usr/local/hadoop-3.3.1/share/hadoop/tools/lib/hadoop-streaming-3.3.1.jar \
   -input 20news-18828/rec.sport.baseball/100521 \
   -output output \
   -file mapper.py  -file reducer.py  \
   -mapper 'python mapper.py'  -reducer 'python reducer.py'
```
